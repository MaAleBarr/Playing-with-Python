```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

Create mock data for plots

```python
x=np.arange(20)
y=np.random.normal(10,1,20)
z=np.random.normal(10,2,20)
```

Create your first plot

```python
fig=plt.figure()
ax=fig.add_subplot(111) # num rows, num cols, nun chart
ax.plot(x,y)
```

Create two plots with line and scatterplot

```python
fig=plt.figure()
ax1=fig.add_subplot(121)
ax1.plot(x,y)
ax2=fig.add_subplot(122)
ax2.scatter(y,z)
```

Create two plots with histogram and boxplot

```python
fig=plt.figure()
ax1=fig.add_subplot(121)
ax2=fig.add_subplot(122)
ax1.hist(y)
ax2.boxplot([y,z])
```

Plot multiple data sets on same axes with labels and a legend

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(x,y,label='Thing #1')
ax.plot(x,z, label='Thing #2')
ax.legend()
```

Add titles and suptitle to figure 

```python
fig=plt.figure()
ax1=fig.add_subplot(121)
ax1.plot(x,y)
ax1.set_title("Line")

ax2=fig.add_subplot(122)
ax2.scatter(y,z)
s=y
ax2.set_title("Scatter")
fig.suptitle("Two Charts")
```

Add x and y axis labels with font sizes

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(x,y)
ax.set_ylabel('Y-Axis', fontsize=10)
ax.set_xlabel('X-Axis', fontsize=20)
```

Save your figure as png file

```python
fig.savefig('figure_wSeaborn.png')
```

#### Create a figure with 4 subplots
- Histogram of Titanic age distribution
- Scatterplot of Age and Ticket Fares
- Boxplot of Ages for Survivors vs. Deceased
- Boxplot of Ages for Men vs. Women
- Save figure as a png file


```python
data=pd.read_csv('train.csv')
data.head()
```

```python
data.Age
```

```python
age_clean=data.Age.fillna(value=30)
fig=plt.figure()
ax1=fig.add_subplot(221)
ax1.hist(age_clean)
ax1.set_xlabel('Age', fontsize=16)
ax1.set_ylabel('Number of ppl', fontsize=16)

ax2=fig.add_subplot(222)
ax2.scatter(age_clean,data.Fare)
ax2.set_xlabel('Age', fontsize=16)
ax2.set_ylabel('Fare', fontsize=16)

ax3=fig.add_subplot(223)
surv=data.Age[data.Survived==1].fillna(value=30)
dead=data.Age[data.Survived==0].fillna(value=30)
ax3.boxplot([surv,dead])
ax3.set_xlabel('', fontsize=16)
ax3.set_ylabel('Age', fontsize=16)

ax4=fig.add_subplot(224)
female=data.Age[data.Sex=='female'].fillna(value=30)
male=data.Age[data.Sex=='male'].fillna(value=30)
ax4.boxplot([female,male])
ax4.set_xlabel('', fontsize=16)
ax4.set_ylabel('Age', fontsize=16)
```

```python
fig.savefig('4plotfigure.png')
```

```python
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2
speed = np.sqrt(U*U + V*V)

plt.streamplot(X, Y, U, V, color=U, linewidth=2, cmap=plt.cm.autumn)
plt.colorbar()

f, (ax1, ax2) = plt.subplots(ncols=2)
ax1.streamplot(X, Y, U, V, density=[0.5, 1])

lw = 5*speed/speed.max()
ax2.streamplot(X, Y, U, V, density=0.6, color='k', linewidth=lw)

plt.show()
```
